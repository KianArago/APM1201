```python
import random
import time
import matplotlib.pyplot as plt
import numpy as np

def measure_sort_time(size):
    random_list = [random.randint(0, 100000) for _ in range(size)]

    start_time = time.time()
    sorted_list = sorted(random_list)
    end_time = time.time()

    return end_time - start_time

def experimental_analysis(max_size, step):
    sizes = []
    times = []

    for size in range(step, max_size + 1, step):
        elapsed_time = measure_sort_time(size)
        sizes.append(size)
        times.append(elapsed_time)
        print(f"Size: {size}, Time: {elapsed_time:.6f} seconds")

    return sizes, times

def plot_results(sizes, times):
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, times, marker='o', linestyle='-', color='b', label='Measured Time')

    log_sizes = np.log2(sizes)
    fit = np.polyfit(log_sizes, times, 1)
    plt.plot(sizes, np.polyval(fit, log_sizes), 'r--', label='O(n log n) Fit')

    plt.xlabel('Input Size (n)')
    plt.ylabel('Time (seconds)')
    plt.title('Experimental Analysis of Python Sorted Method')
    plt.legend()
    plt.grid(True)
    plt.show()

max_size = 10000
step = 1000
sizes, times = experimental_analysis(max_size, step)

plot_results(sizes, times)
```
